# TASK-1:Exploratory Data Analysis

```python
# @title Importing Required Libraries
#In this task we are performing EDA,these are the required libraries to perform EDA.
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# @title Reading The Black Friday Sales Dataset.
# Reading the csv file
Sales = pd.read_csv("Black Friday Sales.csv")
# Here we are displaying the top 5 rows.
Sales.head(5)
```

```python
# Here we are displaying the last 5 rows.
Sales.tail(5)
```

```python
# @title Checking The DataTypes
# Using the dtypes command we are checking the types of data.
Sales.dtypes
```

```python
# @title Dropping The Irrelevant Columns
#Here we are dropping the irrelevant columns, In my case i am dropping the Product_ID
Sales = Sales.drop(['Product_ID'], axis=1)
# Here we are displaying the top 5 rows
Sales.head(5)
```

```python
# @title Dropping The Duplicate Rows
# Here we are finding the duplicate rows using the duplicated() command.
rows_duplicate = Sales[Sales.duplicated()]
# Printing the duplicate rows.
print("Number of duplicate rows: ", rows_duplicate.shape)
```

```python
#Here i am counting the number of rows.
Sales.count()
```

```python
# Here we are dropping the duplicates rows from the dataframe using the drop_duplicates() command.
Sales = Sales.drop_duplicates()
# Here we are displaying the top 5 rows.
Sales.head(5)
```

```python
# After dropping the duplicate rows again i am counting the number of rows.
Sales.count()
```

```python
# @title Replacing The Missing Data With Numeric Values
# In the 'Gender' column we are Replacing 'F' with 0 and 'no' with 1 using replace command.
Sales['Gender'] = Sales['Gender'].replace({'F': 0, 'M': 1})

# Here we are displaying the dataFrame with the replaced values of 'Gender' column.
print(Sales)
```

```python
# In the 'Age' column we are replacing 0-17': 1, '18-25': 2, '26-35': 3, '36-45': 4, '46-50': 5, '51-55': 6, '55+': 7 using replace command.
Sales['Age'] = Sales['Age'].replace({'0-17': 1, '18-25': 2, '26-35': 3, '36-45': 4, '46-50': 5, '51-55': 6, '55+': 7})

# Here we are displaying the dataFrame with the replaced values of 'Age' column.
print(Sales)
```

```python
# In the 'Stay_In_Current_City_Years' column we are Replacing '+':'' using str.replace() command.
Sales['Stay_In_Current_City_Years'] = Sales['Stay_In_Current_City_Years'].str.replace('+','')

# Here we are displaying the dataFrame with the replaced values of 'Stay_In_Current_City_Years' column.
print(Sales)
```

```python
# @title Filling The Missing Values
# Using the isnull() command i am checking the missing values in the dataset
print(Sales.isnull().sum())
```

```python
# We are filling the missing values in Product_Category_2 with the mode value of Product_Category_2.
Sales["Product_Category_2"].fillna(Sales["Product_Category_2"].mode()[0],inplace=True)
# Here we are displaying the dataFrame after filling the missing values of 'Product_Category_2' column.
print(Sales)
```

```python
# We are filling the missing values in Product_Category_3 with the mode value of Product_Category_3.
Sales["Product_Category_3"].fillna(Sales["Product_Category_3"].mode()[0],inplace=True)
# Here we are displaying the dataFrame after filling the missing values of 'Product_Category_3' column.
print(Sales)
```

```python
# After filling the missing values, again i am checking is there any other null values  in the dataset using isnull() command
print(Sales.isnull().sum())
```

```python
# @title Calculating InterQuartile Range
# Calculating q1 which is 25%
q1 = Sales.quantile(0.25)
# Calculating q2 which is 75%
q2 = Sales.quantile(0.75)
# By subtracting q3-q2,here we calculating interquartile range
InterQuartileRange = q2 - q1
# Printing the Interquartile range
print(InterQuartileRange)
```

```python
# Here we are Removing Outliers
Sales = Sales[~((Sales < (q1 - 1.5 * InterQuartileRange)) |(Sales > (q2 + 1.5 * InterQuartileRange))).any(axis=1)]
# Printing the shape
Sales.shape
```

# TASK-2:Feature Selection

```python
# @title Label Encoder

# Importing the required libraries from the scikit-learn
from sklearn.preprocessing import LabelEncoder
# Here we are intiallizing the label encoder to encode the categorical data
label_encoder = LabelEncoder()
# Here we are applying the label encoder to "City_Category" columns and fit_transform command transforms the categorical values into numerical values
Sales['City_Category'] = label_encoder.fit_transform(Sales['City_Category'])
```

```python
# @title Feature Selection-SelectKBest
import pandas as pd
from sklearn.feature_selection import SelectKBest

# Here we are separating the target variable and features,
X = Sales.drop('User_ID', axis=1)
y = Sales['Product_Category_1']

# Here we are creating feature selection technique,we are instantiating feature selection method as selectKBest-it will select the most relevant features
Feature = SelectKBest(k=7)  # We kept k as 7 to select the top 7 features(K=7)

# To know the most important features i am using the fit() method to train the data
Feature.fit(X, y)

# Here we are using the get_support() command to get the selected features
most_important_features = Feature.get_support()
most_important_features_names = X.columns[most_important_features]

# Here we are printing the most important features
print("Most Important Features:", most_important_features_names)
```

# TASK-3-Model Selection & Model Training

```python
# @title Importing Required libraries
# Importing numpy and pandas libraries to read the data
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
import sklearn.metrics as metrics
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
```

```python
# Here we are Creating X and y
X = Sales['Purchase']
y = Sales['Product_Category_1']
```

```python
# @title Creating The Train And Test Cases
# Here we are splitting the varaibles as training_size as 70 and testing_size as 30
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
# @title Training Data
# To verify the split we are printing the shapes
print(X_train)
print(y_train)
```

```python
# @title Testing Data
# To verify the split we are printing the shapes
print(X_test)
print(y_test)
```

```python
# @title Training The Model-Guassian Model
# Here X_train data is 1D array to perform the Gaussian Modle data should be in 2D array. So we are reshaping the X_train data using values.reshape() command
X_train = X_train.values.reshape(-1, 1)
# Here we are creating Gaussian model
Gaussian_model = GaussianNB()
# We are fitting the Guassian model Using training data
Gaussian_model.fit(X_train, y_train)
```

# Task-4:Model Evaluation

```python
# @title Evaluating The Model
# Here X_test data is 1D array to perform the Gaussian Model data should be in 2D array. So we are reshaping the x_test data using values.reshape() command
X_test = X_test.values.reshape(-1, 1)
# Here we are using the Gaussian model to make the predictions on the data which is reshaped
y_pred = Gaussian_model.predict(X_test)
# Here we are printing the y_predicted values
print(y_pred)
```

```python
# @title Finding The Accuracy
# Here we are comparing y_test,y_pred using accuracy_score()
accuracy = accuracy_score(y_test, y_pred)*100
# Here we are printing the accuracy of the testing data
print(f"Accuracy: {accuracy}")
```

```python
# @title Creating A Confusion Matrix
# Here we are comparing y_test,y_pred using confusion_matrix()
conf_matrix = confusion_matrix(y_test, y_pred)
# Here we are displaying the confusion matrix
print("Confusion Matrix:")
print(conf_matrix)
```

```python
# @title Decision Tree Classifier Results
# Here we are printing the classification report
print("Classification Report:")
# Here we are comparing y_test,y_pred using classification_report()
print(classification_report(y_test, y_pred))
```

# TASK-3(1)-Model Selection & Model Training



```python
# @title Importing Required libraries
# Importing numpy and pandas libraries to read the data
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
import sklearn.metrics as metrics
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
```

```python
# Here we are Creating X and y
X = Sales['Purchase']
y = Sales['Product_Category_1']
```

```python
# @title Creating The Train And Test Cases
# Here we are splitting the varaibles as training_size as 70 and testing_size as 30
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7,
                                                    test_size = 0.3, random_state = 100)
```

```python
# @title Training Data
# To verify the split we are printing the shapes
print(X_train)
print(y_train)
```

```python
# @title Testing Data
# To verify the split we are printing the shapes
print(X_test)
print(y_test)
```

```python
# @title Training The Model-Decision Tree Classifier
# Here X_train data is 1D array to perform the decision tree classifier data should be in 2D array. So we are reshaping the X_train data using values.reshape() command
X_train = X_train.values.reshape(-1, 1)
# Here we are creating Decision Tree model
dtree_model = DecisionTreeClassifier()
# We are fitting the Decision Tree model Using training data
dtree_model.fit(X_train, y_train)
```

# Task-4(1):Model Evaluation

```python
# @title Evaluating The Model
# Here X_test data is 1D array to perform the decision tree classifier data should be in 2D array. So we are reshaping the x_test data using values.reshape() command
X_test = X_test.values.reshape(-1, 1)
# Here we are using the decision tree model to make the predictions on the data which is reshaped
y_pred = dtree_model.predict(X_test)
# Here we are printing the y_predicted values
print(y_pred)
```

```python
# @title Finding The Accuracy
# Here we are comparing y_test,y_pred using accuracy_score()
accuracy = accuracy_score(y_test, y_pred)*100
# Here we are printing the accuracy of the testing data
print(f"Accuracy: {accuracy}")
```

```python
# @title Creating A Confusion Matrix
# Here we are comparing y_test,y_pred using confusion_matrix()
conf_matrix = confusion_matrix(y_test, y_pred)
# Here we are displaying the confusion matrix
print("Confusion Matrix:")
print(conf_matrix)
```

```python
# @title Decision Tree Classifier Results
# Here we are printing the classification report
print("Classification Report:")
# Here we are comparing y_test,y_pred using classification_report()
print(classification_report(y_test, y_pred))
```
